```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import ensemble
from sklearn.decomposition import PCA
from scipy import stats
from matplotlib import rcParams
from sklearn.preprocessing import normalize
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from matplotlib import rcParams
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV


import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
sns.set_style('whitegrid')
rcParams['figure.figsize'] = 10,7
```

```python
df = pd.read_csv('cardio_train.csv', sep=';')
```

```python
df.head()
```

```python
df.shape
```

```python
df['age'] = round(df['age'] / 365, 1)
df['mean arterial pressure'] = (df['ap_hi'] + df['ap_lo']) / 2
```

```python
df.head()
```

```python
X = df.drop(['id', 'cardio'],1)
y = df['cardio']
```

```python
logreg = LogisticRegression()
c_range = range(1,1000,10)
grid_params = dict(penality=['l2', 'l1'], C=c_range)
grid = GridSearchCV(logreg, grid_params, X, y, cv=10, scoring='accuracy')
```

```python
knn = KNeighborsClassifier(n_neighbors=5, weights='distance')

# Normalize data for Euclidean geometry
norm_X = normalize(X)
k_range = list(range(1,60))
weighted_options = ['uniform', 'distance']
grid_params = dict(n_neighbors=k_range, weights=weighted_options)
grid = GridSearchCV(knn, grid_params, cv=10, scoring='accuracy', return_train_score=False)
grid.fit(norm_X, y)
```

```python
pd.DataFrame(grid.cv_results_)[['mean_test_score', 'std_test_score', 'params']]
```
